```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Population of India.csv")
df.head(3)
```

```python
import matplotlib.pyplot as plt

# Plot the chart
df[['Male', 'Female']].plot(kind='barh', figsize=(12, 8), width=0.8)

# Customize the chart
plt.title('Male vs Female Population Distribution by State/UT')
plt.xlabel('Population')
plt.ylabel('State/UT')
plt.legend(title='Gender')

# Adjust layout to avoid unwanted spacing (tail)
plt.tight_layout()

# Show the plot
plt.show()
```
